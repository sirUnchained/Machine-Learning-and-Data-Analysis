## 2.4.1 analysis result

1. **dataset Overview**:
    - dataset contains 11 columns and 234 rows
    - dataset contains both numerical and categorical variables

2. **keys**:
    - *manufacturer*: the producer company name
    - *model*: the car model 
    - *displ*: car engine capacity (displacement in liters)
    - *year*: year of manufacture of the car
    - *cyl*: number of car cylinders
    - *trans*: is the car auto or manual? (transmission type)
    - *drv*: drive train type (f = front-wheel drive, r = rear-wheel drive, 4 = 4-wheel drive)
    - *cty*: distance the car travels in the city with one gallon of gasoline (miles per gallon)
    - *hwy*: distance the car travels on the highway with one gallon of gasoline (miles per gallon)
    - *fl*: car fuel type (e = ethanol, d = diesel, r = regular, p = premium, c = CNG)
    - *class*: car class/type (compact, midsize, SUV, 2seater, minivan, pickup, subcompact)
